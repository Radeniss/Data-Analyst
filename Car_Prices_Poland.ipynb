```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
```

```python
model_counts = df['model'].value_counts().head(10)

# Ukuran plot
plt.figure(figsize=(10, 6))

# Membuat bar plot
plt.bar(model_counts.index, model_counts.values, color='skyblue')

# Label sumbu X dan Y
plt.xlabel('Model')
plt.ylabel('Frekuensi')

# Judul plot
plt.title('Frekuensi Kemunculan Model')

# Menampilkan plot
plt.show()
```
